```python
# Identifying categorical and numerical columns
categorical_cols = fmp_data.select_dtypes(include=['object']).columns.tolist()
numerical_cols = fmp_data.select_dtypes(include=['number']).columns.tolist()

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

**Exploratory Data Analysis**

```python
import matplotlib.pyplot as fmp_plt
import seaborn as fmp_sns
import warnings as fmp_wgs
fmp_wgs.filterwarnings('ignore')
```

**Distribution of Goals Scored**

```python
fmp_plt.figure(figsize=(12, 6))
fmp_sns.histplot(fmp_data['FTHG'], bins=5, kde=True, color='blue', label='Home Goals')
fmp_sns.histplot(fmp_data['FTAG'], bins=5, kde=True, color='red', label='Away Goals')
fmp_plt.title('Distribution of Full-Time Goals')
fmp_plt.xlabel('Goals')
fmp_plt.ylabel('Frequency')
fmp_plt.legend()
fmp_plt.show()
```

**Team Goals vs Points**

```python
fmp_plt.figure(figsize=(12, 6))
fmp_sns.scatterplot(x='HTGS', y='HTP', data=fmp_data, color='blue', label='Home Team Goals vs Points')
fmp_sns.scatterplot(x='ATGS', y='ATP', data=fmp_data, color='red', label='Away Team Goals vs Points')
fmp_plt.title('Team Goals vs. Points')
fmp_plt.xlabel('Goals Scored')
fmp_plt.ylabel('Points')
fmp_plt.legend()
fmp_plt.show()
```

**Goal Differences**

```python
fmp_plt.figure(figsize=(12, 6))
fmp_sns.histplot(fmp_data['HTGD'], bins=5, kde=True, color='blue', label='Home Goal Difference')
fmp_sns.histplot(fmp_data['ATGD'], bins=5, kde=True, color='red', label='Away Goal Difference')
fmp_plt.title('Distribution of Goal Differences')
fmp_plt.xlabel('Goal Difference')
fmp_plt.ylabel('Frequency')
fmp_plt.legend()
fmp_plt.show()
```

**Points Difference vs. Form Points Difference**

```python
fmp_plt.figure(figsize=(12, 6))
fmp_sns.scatterplot(x='DiffPts', y='DiffFormPts', data=fmp_data, color='green')
fmp_plt.title('Points Difference vs. Form Points Difference')
fmp_plt.xlabel('Points Difference')
fmp_plt.ylabel('Form Points Difference')
fmp_plt.show()
```

**Outlier Detection**

```python
# Set up the matplotlib figure
fig, axes = fmp_plt.subplots(nrows=len(numerical_cols), ncols=3, figsize=(18, 4*len(numerical_cols)))

# Plotting
for i, col in enumerate(numerical_cols):
    # Box plot
    fmp_sns.boxplot(data=fmp_data[col], ax=axes[i, 0])
    axes[i, 0].set_title(f'Box Plot of {col}')

    # Violin plot
    fmp_sns.violinplot(data=fmp_data[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Violin Plot of {col}')

    # Distribution plot
    fmp_sns.histplot(data=fmp_data[col], kde=True, ax=axes[i, 2])
    axes[i, 2].set_title(f'Distribution Plot of {col}')

fmp_plt.tight_layout()
fmp_plt.show()
```
