**Outlier Detection**

```python
# Set up the matplotlib figure
fig, axes = fmp_plt.subplots(nrows=len(numerical_cols), ncols=3, figsize=(18, 4*len(numerical_cols)))

# Plotting
for i, col in enumerate(numerical_cols):
    # Box plot
    fmp_sns.boxplot(data=fmp_data[col], ax=axes[i, 0])
    axes[i, 0].set_title(f'Box Plot of {col}')

    # Violin plot
    fmp_sns.violinplot(data=fmp_data[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Violin Plot of {col}')

    # Distribution plot
    fmp_sns.histplot(data=fmp_data[col], kde=True, ax=axes[i, 2])
    axes[i, 2].set_title(f'Distribution Plot of {col}')

fmp_plt.tight_layout()
fmp_plt.show()
```

**Removing Outliers**

```python
numerical_cols = ['FTHG','FTAG','HTGS','ATGS','HTGC','ATGC','HTP',
 'ATP','MW','HTFormPts','ATFormPts','HTGD','ATGD','DiffPts','DiffFormPts']
```

```python
# Function to remove outliers and update fmp_data
def remove_outliers_and_update(fmp_dat, col):
    Q1 = fmp_dat[col].quantile(0.25)
    Q3 = fmp_dat[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR
    fmp_dat[col] = fmp_dat[col][(fmp_dat[col] >= lower_fence) & (fmp_dat[col] <= upper_fence)]
    return fmp_dat

# Set up the matplotlib figure
fig, axes = fmp_plt.subplots(nrows=len(numerical_cols), ncols=3, figsize=(18, 4*len(numerical_cols)))

# Plotting
for i, col in enumerate(numerical_cols):
    # Remove outliers and update fmp_data
    fmp_data = remove_outliers_and_update(fmp_data, col)

    # Box plot
    fmp_sns.boxplot(data=fmp_data[col], ax=axes[i, 0])
    axes[i, 0].set_title(f'Box Plot of {col}')

    # Violin plot
    fmp_sns.violinplot(data=fmp_data[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Violin Plot of {col}')

    # Distribution plot
    fmp_sns.histplot(fmp_data[col], kde=True, ax=axes[i, 2])
    axes[i, 2].set_title(f'Distribution Plot of {col}')

fmp_plt.tight_layout()
fmp_plt.show()
```
